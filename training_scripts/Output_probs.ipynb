```python
import numpy as np
import torch
file = './results/train'

log_p1_0 = np.load(file+'/log_p1_0.npy')
log_p1_1 = np.load(file+'/log_p1_1.npy')
log_p1_2 = np.load(file+'/log_p1_2.npy')
# log_p1_3 = np.load(file+'/log_p1_3.npy')

log_p1 = np.concatenate((log_p1_0,log_p1_1,log_p1_2))

log_p2_0 = np.load(file+'/log_p2_0.npy')
log_p2_1 = np.load(file+'/log_p2_1.npy')
log_p2_2 = np.load(file+'/log_p2_2.npy')
# log_p2_3 = np.load(file+'/log_p2_3.npy')

log_p2 = np.concatenate((log_p2_0,log_p2_1,log_p2_2))

pp = np.exp(log_p2 - log_p1)
p1 = 1/(pp+1)
p2 = pp/(pp+1)

    
r1_0 = np.load(file+'/r1_0.npy')
r1_1 = np.load(file+'/r1_1.npy')
r1_2 = np.load(file+'/r1_2.npy')
# r1_3 = np.load(file+'/r1_3.npy')

r1 = np.concatenate((r1_0,r1_1,r1_2))

r2_0 = np.load(file+'/r2_0.npy')
r2_1 = np.load(file+'/r2_1.npy')
r2_2 = np.load(file+'/r2_2.npy')
# r2_3 = np.load(file+'/r2_3.npy')

r2 = np.concatenate((r2_0,r2_1,r2_2))

# r = np.array(r1,r2)
# print(np.size(r))
rr = np.exp(r2 - r1)
pr1 = 1/(rr+1)
pr2 = rr/(rr+1)

import matplotlib.pyplot as plt

plt.figure(figsize=[3,3])
plt.title('Reference Model')
plt.xlabel('Preference on $y$')
plt.hist(p1,100,range = [0,1])
# plt.hist(pr1,100)
plt.grid()
plt.legend()
plt.savefig('refmodel1.3.pdf',bbox_inches='tight')

plt.figure(figsize=[3,3])
plt.title('Reward Model')
plt.xlabel('Preference on $y$')
# plt.hist(p1,100,range = [0,1])
plt.hist(pr1,100)
plt.grid()
plt.legend()
plt.savefig('rewardmodel1.3.pdf',bbox_inches='tight')
```
